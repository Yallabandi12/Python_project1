```python
from prophet import Prophet


# Group data to get number of cars sold per day
daily_sales = df.groupby('Date Sold').size().reset_index(name='y')

# Prophet requires columns 'ds' and 'y'
daily_sales.rename(columns={'Date Sold': 'ds'}, inplace=True)

# Initialize and fit Prophet model
model = Prophet()
model.fit(daily_sales)

# Create future dates (e.g., next 6 months)
future = model.make_future_dataframe(periods=180, freq='D')  # 'D' = daily

# Forecast
forecast = model.predict(future)

# Plot forecast
fig1 = model.plot(forecast)
plt.title("Forecast of Car Sales")
plt.xlabel("Date")
plt.ylabel("Cars Sold")
plt.show()

# Optional: Plot trend/seasonality components
fig2 = model.plot_components(forecast)
plt.show()
```

```python
# Step 2: Group data by month and count number of cars sold
monthly_sales = df.resample('M', on='Date Sold').size().reset_index(name='y')

# Step 3: Prophet expects 'ds' and 'y' columns
monthly_sales.rename(columns={'Date Sold': 'ds'}, inplace=True)

# Step 4: Initialize and fit the model
model = Prophet()
model.fit(monthly_sales)

# Step 5: Create future months (e.g., forecast next 12 months)
future = model.make_future_dataframe(periods=6, freq='M')

# Step 6: Forecast
forecast = model.predict(future)

# Step 7: Plot the forecast
fig1 = model.plot(forecast)
plt.title("Monthly Car Sales Forecast")
plt.xlabel("Date")
plt.ylabel("Cars Sold")
plt.show()

# Step 8 (Optional): Decompose trend/seasonality
fig2 = model.plot_components(forecast)
plt.show()
```

```python
import statsmodels.api as sm


df = df.sort_values('Date Sold')

# Step 2: Resample to get monthly sales count
monthly_sales = df.resample('M', on='Date Sold').size()
monthly_sales.name = 'Sales Count'

# Step 3: Plot the data (optional but useful)
monthly_sales.plot(title='Monthly Sales Count', figsize=(12,6))
plt.ylabel('Number of Sales')
plt.show()

# Step 4: Fit SARIMA model (example parameters)
model = sm.tsa.statespace.SARIMAX(monthly_sales,
                                  order=(1,1,1),
                                  seasonal_order=(1,1,1,12),  # monthly seasonality
                                  enforce_stationarity=False,
                                  enforce_invertibility=False)
results = model.fit()

# Summary of model
print(results.summary())

# Step 5: Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_index = pd.date_range(start=monthly_sales.index[-1] + pd.offsets.MonthBegin(),
                               periods=12, freq='M')
forecast_series = pd.Series(forecast.predicted_mean.values, index=forecast_index)

# Step 6: Plot forecast
plt.figure(figsize=(12,6))
plt.plot(monthly_sales, label='Historical Monthly Sales')
plt.plot(forecast_series, label='Forecasted Sales', color='red')
plt.title('Monthly Sales Forecast (SARIMA)')
plt.ylabel('Number of Sales')
plt.legend()
plt.show()
```
